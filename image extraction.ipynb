```python
import os
import io
from pdfplumber import open as pdf_open
from PIL import Image
from PIL import UnidentifiedImageError

# Create the directory if it doesn't exist
directory = "fr6"
if not os.path.exists(directory):
    os.makedirs(directory)

# Open the PDF file
with pdf_open("6eme fr.pdf") as pdf:
    for page_num in range(len(pdf.pages)):
        page = pdf.pages[page_num]
        for img_num, img in enumerate(page.images):
            try:
                image_data = img["stream"].get_data()
                image = Image.open(io.BytesIO(image_data))
                width, height = image.size
                if height > 170 and width > 170:
                    filename = f"{directory}/image_{page_num}_{img_num}.jpg"
                    with open(filename, "wb") as fp:
                        fp.write(image_data)
            except UnidentifiedImageError:
                print(f"Skipping image {page_num}_{img_num} - UnidentifiedImageError")
```

```python
import PyPDF2

def extract_text_from_pdf(file_path):
    with open(file_path, 'rb') as file:
        pdf_reader = PyPDF2.PdfReader(file)
        text = ''
        for page_num in range(len(pdf_reader.pages)):
            page = pdf_reader.pages[page_num]
            text += page.extract_text()
    return text

# Exemple d'utilisation
pdf_file = '6eme fr.pdf'
extracted_text = extract_text_from_pdf(pdf_file)
print(extracted_text)
```
